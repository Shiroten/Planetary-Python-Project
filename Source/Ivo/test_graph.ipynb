```python
from universe import Oschi
from universe import Universe
import matplotlib.pyplot as plt
% matplotlib inline

sun = Oschi(
    "Sun",
    (0, 0, 0),
    1.9889 * 10 ** 30,
    (0, 0, 0),
    0.2
)
earth = Oschi(
    "Earth",
    (149_597_890_000, 0, 0),
    5.97219 * 10 ** 24,
    (0, -29_786, 0),
    0.1
)

universe = Universe(60 * 60, [sun, earth])
result_e = []
for _ in range(364 * 60):
    os_list = universe.simulate()
    os = os_list[1]
    result_e.append(os.position)
plt.figure(figsize=(10, 10))
plt.plot([x[0] for x in result_e], [x[1] for x in result_e])
```
